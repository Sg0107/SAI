```python
from get_crypto_data import get_data
from datetime import datetime
from Indicators import bollinger_bands, RSI , Ema
import pandas as pd
from pre_process import pre_processing
import numpy as np
import tensorflow as tf
```

```python
# Extract from binance api
start_time = datetime(2019,1,1,0,0)
end_time = datetime(2024,11,30,23,59)
time_frame = ['1h', '30m','15m','5m']
symbol = 'BTC/USDT'
sym = 'BTC_USDT'
pth = []
for t_f in time_frame:
    path = get_data(start_time, end_time, t_f, symbol,sym)
    pth.append(path)
```

```python
# Calculate indicator values and store to csv files with name with_indicator
for path in pth: 
    df = pd.read_csv(path)
    df = RSI(df,14)
    df = bollinger_bands(df,20,1.94937)
    df = Ema(df,9)
    df = Ema(df,20)
    df = Ema(df,200)
    df.to_csv(f"with_Indicator_{path}")
```
